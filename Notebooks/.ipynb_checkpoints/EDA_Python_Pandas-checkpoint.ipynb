```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/media/zmwaris/mydrive/databases and datasets/train_V2.csv")
pd.set_option("display.max_columns", 30)
```

### Let's explore the columns:
**Id** - Unique Id for each player \
**groupId** - Id to identify the group in which the player has participated. Changes everytime with a particular match even if the players in the group remains same. \
**matchId** - To identify individual matches of the player. \
**assists** - Assist provided by the player while in a group. \
**boosts** - Number of boost items used by the player. \
**damageDealt** - total damage dealt. \
**DBNOs** - Number of enemy players knocked in the match. \
**headshotKills** - Number of players killed by headshots. \
**heals** - Number of healing items used \
**killPlace** - Ranking of player on the kill leaderboard \
**killPoints** - Kills based ranking of a player \
**kills** - Number of enemy players killed. \
**killStreaks** - Most number of players killed in a duration \
**longestKill** - A kill where the distance between player and killed enemy is longest. \
**matchDuration** - Duration of a match \
**matchType** - Type of match(Solo, Duo, Squad etc.) \
**maxPlace** - Placement of player in the match \
**numGroups** - Number of groups we have data for in the match \
**rankPoints** - Points gained based on ranking \
**revives** - Number of revives given \
**rideDistance** - Total distance travelled by vehicles. \
**roadKills** - Number of kills while in a vehicle. \
**swimDistance** - Total distance the player swan in the match. \
**teamKills** - Number of times player killed during friendly-fire. \
**vehicleDestroys** - Number of vehicle destroyed. \
**walkDistance** - Total distance travelled by running \
**weaponsAcquired** - Number of weapns picked up. \
**winPoints** - Elo points of player based on wins. \
**winPlacePerc** - Probablility of winning for the player in the match \

#### Let's explore the data

```python
avg = data["kills"].mean()
maxK = data["kills"].max()
quan = data["kills"].quantile(0.99)

print(f"On an average a player has {avg:.4f} kills. Maximum number of kills ever to happend is {maxK} and 99% of players have kills less than or equal to {quan}.")
```

#### Plotting the kills counts

```python
data[data["kills"]>quan].count()
```

```python
sns.countplot(data=data, x="kills")
```

```python
df = data.copy()
```

```python
df.loc[df['kills'] > df['kills'].quantile(0.99)] = '8+'
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df['kills'].astype('str').sort_values())
plt.title("Kill Count",fontsize=15)
plt.show()
```

```python
noKills = df[df['kills']==0]
plt.figure(figsize=(15, 12))
plt.title('Damage Dealt with no kills')
sns.distplot(x=noKills['damageDealt'])
plt.show()
```

We could see that most people who has 0 kills also have little to no damage dealt to enemy players.

```python
perc = 100*len(noKills[noKills["winPlacePerc"]==1])/len(data)
print("{} players ({:.4f}%) have won a game without a single kill".format(len(noKills[noKills["winPlacePerc"]==1]), perc))
```

```python
noDamage = data[data["damageDealt"]==0]
wonNoDamage = noDamage[noDamage["winPlacePerc"]==1]
percentage = len(wonNoDamage)/len(data)
```

```python
print("{} players ({:.4f})% have won the game with 0 damage".format(len(wonNoDamage), percentage))
```

```python
sns.jointplot(data=data, x='winPlacePerc', y='kills', color='b')
```

We see that as the number of kills increase the winning probability also increases.
