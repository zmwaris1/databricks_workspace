```python
import pandas as pd
```

```python
dest = pd.read_csv('/media/zmwaris/mydrive/databases and datasets/packt/dest.csv')
tips = pd.read_csv('/media/zmwaris/mydrive/databases and datasets/packt/tips.csv')
```

```python
dest.head()
```

```python
tips.head()
```

```python
# Joins in Pandas
```

```python
# Inner join
df_inner = pd.merge(dest, tips, how='inner', on='EmpNr')
df_inner
```

```python
df_outer = pd.merge(dest, tips, how='outer', on='EmpNr')
df_outer
```

```python
df_left = pd.merge(tips, dest, on='EmpNr', how='left')
df_left
```

```python
df_right = pd.merge(tips, dest, on='EmpNr', how='right')
df_right
```
